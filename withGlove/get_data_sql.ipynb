```python
import pandas as pd
def product_info_by_bucket(): 
    '''
    Description: collect product information(name, description, url, image_url, attribute and values) for products in a terminmal buckets
        the output is pivited to have the columns by attribute_name and the cells to contain the value_name.
    ############
    params:
        {bucket_name: str} bucket name from the global taxonomy how it appears in console
            it converts this to lower case to ensure all values are returned
            This is uses a 'LIKE' clause and searches anywhere in the string
    '''
    def connection():
        '''
        Desctipion: establish connection to sandbox database
        '''
        import psycopg2 
        feeddata = {
            "database":"feeddata",
            "user":"groupby",
            "password":"gbi123gbi",
            "host":"sandbox.edgecaseprod.com",
            "port":"5432"
        }
        conn = psycopg2.connect(**feeddata)
        return conn
#     bucket_name = bucket_name.lower()
    query = ("""
SELECT
   DISTINCT
   pb.bucket_id
   ,p.external_id
   ,p.customer_id
   ,b.bucket_name
   ,p.product_id
   ,p.product_name
   ,p.description
   ,p.image_url
   ,p.customer_id
FROM products_buckets AS pb
INNER JOIN (
       SELECT
           id as product_id,
           external_id,
           customer_id
           ,name as product_name
           ,description
           ,image_url
       FROM products
   ) AS p
ON pb.product_id = p.product_id
INNER JOIN (
       SELECT
           id AS bucket_id
           ,name AS bucket_name
           ,source
       FROM buckets
       where 6595 = ANY(parent_hierarchy) --V2
       and 6595 = ANY(parent_hierarchy)) as b 
ON pb.bucket_id = b.bucket_id
""")
    #call query
    print('extracting {} buckets from the database')
    df = pd.read_sql(query, connection())
    return df
df = product_info_by_bucket()
```
